```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('BoxPlotStats2.0.csv')

# Header names
Absolute = data["Absolute"]
Intensity = data["Intensity"]
BaselineNoise = data["BaselineNoise"]
Concentricity = data["Concentricity"]
Camber = data["Camber"]
ThicknessWall = data["ThicknessWall"]
TubeType = data["TubeType"]
Freq = data["Frequency"]
Quality = data["Quality"]
TubeTypeNum = data["TubeTypeNum"]
Cost = data["Cost"]
SN = data["SN"]
PeakWidth = data["PeakWidth"]
DiameterOuter = data["DiameterOuter"]
MOEO = data["MOEO"]
DiameterInner = data["DiameterInner"]
MOEI = data["MOEI"]
SameTube = data["SameTube"]
QualityTube = data["QualityTube"]
SameRun = data["SameRun"]
CostvsQuality = data["CostvsQuality"]

def GraphNormal():
    plt.scatter(x,y, s=50,c=TubeTypeNum, cmap="gist_rainbow", edgecolor = "black", linewidth=1, alpha=0.75)
    plt.title(xstr + " vs "+ ystr)
    plt.xlabel(xstr)
    plt.ylabel(ystr)
    cbar = plt.colorbar()
    cbar.set_label('Tube Type')
    plt.grid(color = 'lightgrey', linestyle = '--', linewidth = 0.5)
    plt.show()
    print("\n\n")

def GraphQuality():
    plt.scatter(x,y, 
                s=50, 
                c=Quality,
                cmap="gist_rainbow",
                edgecolor = "black",
                linewidth=1,
                alpha=0.75)
    plt.title(xstr + " vs "+ ystr)
    plt.xlabel(xstr)
    plt.ylabel(ystr)
    cbar = plt.colorbar()
    cbar.set_label('Quality')
    plt.show()
    print("\n\n")
    
def GraphSameRun():
    # Make scatter graph
    plt.scatter(x,y, 
                s=50, 
                c=SameRun,
                cmap="gist_rainbow",
                edgecolor = "black",
                linewidth=1,
                alpha=0.75)
    plt.title(xstr + " vs "+ ystr)
    plt.xlabel(xstr)
    plt.ylabel(ystr)
    cbar = plt.colorbar()
    cbar.set_label('Tube Run')
    plt.grid(color = 'lightgrey', linestyle = '--', linewidth = 0.5)
    plt.show()
    print("\n\n")
    
```

```python
saving = plt.figure()
x, y = TubeType, Quality
xstr, ystr = str("Tube Type"), str("Quality")
GraphNormal()
saving.savefig(xstr + ystr + '.svg', format ='svg', dpi=1200)
```
