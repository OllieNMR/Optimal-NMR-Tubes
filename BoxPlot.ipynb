```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
import numpy as np
import math

data = pd.read_csv('BoxPlotStats2.0.csv')
df = pd.read_csv("Stats.csv")

SquareRoot = math.sqrt(18)

#Heading names     
Absolute = data["Absolute"]
Intensity = data["Intensity"]
BaselineNoise = data["BaselineNoise"]
Concentricity = data["Concentricity"]
Camber = data["Camber"]
ThicknessWall = data["ThicknessWall"]
TubeType = data["TubeType"]
Freq = data["Frequency"]
Quality = data["Quality"]
TubeTypeNum = data["TubeTypeNum"]
Cost = data["Cost"]
SN = data["SN"]
PeakWidth = data["PeakWidth"]
DiameterOuter = data["DiameterOuter"]
MOEO = data["MOEO"]
DiameterInner = data["DiameterInner"]
MOEI = data["MOEI"]
TubeType = data["TubeType"]
SameTube = data["SameTube"]


TextFile = open("BoxPlot.txt", "w")

def known():
    plt.clf()
    xystr = xstr + " vs " + ystr
    slope, intercept, r, p, std_err = stats.linregress(y, x)
    def myfunc(y):
        return slope * y + intercept
    mymodel = list(map(myfunc, y))
    plt.scatter(y, x, color = "grey"), plt.plot(y, mymodel, color = "lightgrey"),plt.grid(),plt.title((xystr) , fontweight = "bold", fontsize = 20), plt.xlabel(xstr , color = "grey"),plt.ylabel(ystr, color = "grey"), plt.show()
    print(xystr)
    print(xystr, file=TextFile)
    print("Slope = ", slope, "\n intercept = ", intercept,"\n r = ", r,"\n p = " , p, "\n standard error", std_err,'\n\n', file=TextFile)
    print("Slope = ", slope, "\n intercept = ", intercept,"\n r = ", r,"\n p = " , p, "\n standard error", std_err)
    
def table():
    aov_table1 = sm.stats.anova_lm(mod, typ=1)
    aov_table2 = sm.stats.anova_lm(mod, typ=2)
    print(mod.summary(), "\n\n", aov_table1, "\n\n", aov_table2, file=TextFile)
    print(mod.summary(), "\n\n", aov_table1, "\n\n", aov_table2)
    print("\n\n", file = TextFile)
    
def Standard():
    standardDev = data.std()
    StandardError = standardDev / SquareRoot
    print("Standard Deviation:\n ", standardDev, "\n", "standard Error:\n", StandardError, "\n",  file = TextFile)
```
